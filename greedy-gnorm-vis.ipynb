```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ===== =====
box_data_path_BERT   = "experiments_results/BERTrandompruned_heads_data_50x25_concated.xlsx"
line_data_path_BERT  = "experiments_results/pruned_heads_accuracy(BERT)basedonGreedyGnorm.csv"

box_data_path_ALBERT  = "experiments_results/ALBERTrandompruned_heads_data_25x13_1group.xlsx"
line_data_path_ALBERT = "experiments_results/pruned_heads_accuracy(ALBERT)basedonGreedyGnorm.csv"

box_data_path_ROBERTA  = "experiments_results/ROBERTArandompruned_heads_data_25x25_1group.xlsx"
line_data_path_ROBERTA = "experiments_results/pruned_heads_accuracy(ROBERTA)basedonGreedyGnorm.csv"

box_data_path_XLM  = "experiments_results/XLM_ROBERTArandompruned_heads_data_25x25_1group.xlsx"
line_data_path_XLM = "experiments_results/pruned_heads_accuracy(XLM_ROBERTA)basedonGreedyGnorm.csv"

# ----------  ----------
def load_box_excel(excel_path):
    """ Excel
     'Heads Pruned'  x x 
     x
    heads(np.ndarray), box_matrix(list[np.ndarray])
    """
    df = pd.read_excel(excel_path)
    if 'Heads Pruned' in df.columns:
        heads = pd.to_numeric(df['Heads Pruned'], errors='coerce').to_numpy()
        value_df = df.drop(columns=['Heads Pruned'])
    else:
        heads = np.arange(len(df))
        value_df = df
    box_matrix = []
    for _, row in value_df.iterrows():
        vals = pd.to_numeric(row, errors='coerce')
        vals = vals[vals.notna()].to_numpy(dtype=float)
        box_matrix.append(vals)
    return heads, box_matrix

def load_line_xy(csv_path):
    """ CSV
    -  'Pruned_Heads'  'Accuracy'  (x, y, df)
    -  y (None, y, df)
    """
    df = pd.read_csv(csv_path)
    # 
    df.columns = [str(c).strip() for c in df.columns]
    cols = set(df.columns.str.lower())

    if 'pruned_heads' in cols and 'accuracy' in cols:
        x = pd.to_numeric(df[[c for c in df.columns if c.lower() == 'pruned_heads'][0]], errors='coerce')
        y = pd.to_numeric(df[[c for c in df.columns if c.lower() == 'accuracy'][0]], errors='coerce')
        m = pd.DataFrame({'x': x, 'y': y}).dropna()
        return m['x'].to_numpy(dtype=float), m['y'].to_numpy(dtype=float), df

    #  y
    best_col = None
    best_count = -1
    for c in df.columns:
        if c.lower() == 'pruned_heads':
            continue
        ser = pd.to_numeric(df[c], errors='coerce')
        cnt = ser.notna().sum()
        if cnt > best_count:
            best_count = cnt
            best_col = c
    y = pd.to_numeric(df[best_col], errors='coerce').dropna().to_numpy(dtype=float)
    return None, y, df

def align_line_to_heads(line_x, line_y, heads):
    """ (line_x,line_y)  heads
    -  line_x  np.interp  x  heads
    -  line_x  heads  y 
    """
    heads = np.asarray(heads, dtype=float)
    if line_x is not None:
        #  & 
        order = np.argsort(line_x)
        x = np.asarray(line_x, dtype=float)[order]
        y = np.asarray(line_y, dtype=float)[order]
        #  x
        x_unique, idx = np.unique(x, return_index=True)
        y_unique = y[idx]
        # 
        return np.interp(heads, x_unique, y_unique, left=y_unique[0], right=y_unique[-1])

    #  x heads  y 
    if len(heads) <= 1:
        return np.full_like(heads, line_y[0], dtype=float)
    # =  heads 0
    total_heads = max(int(np.nanmax(heads)), 1)
    idx = np.round((heads / total_heads) * (len(line_y) - 1)).astype(int)
    idx = np.clip(idx, 0, len(line_y) - 1)
    return line_y[idx]

def get_no_pruning_baseline(line_x, line_y):
    """
    -  line_x  0 x==0  y
    -  y  NaN  y """
    if line_x is not None:
        mask0 = (np.asarray(line_x) == 0)
        if mask0.any():
            return float(np.asarray(line_y)[mask0][0])
    # 
    return float(np.asarray(line_y)[0])

# ---------- 2x2  ----------
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("BERT", "ALBERT", "RoBERTa", "XLM-RoBERTa"),
    horizontal_spacing=0.08,
    vertical_spacing=0.12
)

fig.update_layout(
    template="plotly_white",
    paper_bgcolor="rgba(0,0,0,0)",
    plot_bgcolor="rgba(0,0,0,0)",
    width=1200,
    height=950,
    legend=dict(
        orientation='h',
        yanchor='bottom', y=-0.08,
        xanchor='center', x=0.5
    ),
    margin=dict(l=40, r=20, t=70, b=90)
)

y_range = [0.0, 1.0]

def add_panel(fig, row, col, box_path, line_path, model_name, show_legend=False):
    heads, box_matrix = load_box_excel(box_path)
    line_x, line_y, _ = load_line_xy(line_path)
    line_aligned = align_line_to_heads(line_x, line_y, heads)
    no_pruning = get_no_pruning_baseline(line_x, line_y)

    # Greedy Gnorm
    fig.add_trace(
        go.Scatter(
            x=heads, y=line_aligned,
            mode='lines', name='Greedy Gnorm',
            line=dict(width=2),
            legendgroup='grp', showlegend=show_legend
        ),
        row=row, col=col
    )

    # No Pruning
    fig.add_trace(
        go.Scatter(
            x=[np.min(heads), np.max(heads)],
            y=[no_pruning, no_pruning],
            mode='lines', name='No Pruning',
            line=dict(width=2),
            legendgroup='grp', showlegend=show_legend
        ),
        row=row, col=col
    )

    # Random Pruning 
    for i, h in enumerate(heads):
        if i >= len(box_matrix): break
        fig.add_trace(
            go.Box(
                y=box_matrix[i], x=[h] * len(box_matrix[i]),
                name='Random Pruning',
                boxpoints=False, whiskerwidth=0.6,
                line=dict(width=1),
                legendgroup='grp', showlegend=(show_legend and i == 0)
            ),
            row=row, col=col
        )

    # 
    fig.update_xaxes(title_text="Heads Pruned", row=row, col=col, showgrid=True)
    fig.update_yaxes(title_text="Accuracy", row=row, col=col, showgrid=True, range=y_range)

# 
add_panel(fig, 1, 1, box_data_path_BERT,   line_data_path_BERT,   "BERT",        show_legend=True)
add_panel(fig, 1, 2, box_data_path_ALBERT, line_data_path_ALBERT, "ALBERT",      show_legend=False)
add_panel(fig, 2, 1, box_data_path_ROBERTA, line_data_path_ROBERTA, "RoBERTa",   show_legend=False)
add_panel(fig, 2, 2, box_data_path_XLM,   line_data_path_XLM,     "XLM-RoBERTa", show_legend=False)

fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ==========  ==========
box_data_path_BERT   = "experiments_results/BERTrandompruned_heads_data_50x25_concated.xlsx"
line_data_path_BERT  = "experiments_results/pruned_heads_accuracy(BERT)basedonGreedyGnorm.csv"

box_data_path_ALBERT  = "experiments_results/ALBERTrandompruned_heads_data_25x13_1group.xlsx"
line_data_path_ALBERT = "experiments_results/pruned_heads_accuracy(ALBERT)basedonGreedyGnorm.csv"

box_data_path_ROBERTA  = "experiments_results/ROBERTArandompruned_heads_data_25x25_1group.xlsx"
line_data_path_ROBERTA = "experiments_results/pruned_heads_accuracy(ROBERTA)basedonGreedyGnorm.csv"

box_data_path_XLM  = "experiments_results/XLM_ROBERTArandompruned_heads_data_25x25_1group.xlsx"
line_data_path_XLM = "experiments_results/pruned_heads_accuracy(XLM_ROBERTA)basedonGreedyGnorm.csv"


# ==========  ==========
def load_box_excel(excel_path):
    """ Excel (heads_pruned: np.ndarray, box_matrix: list[np.ndarray])"""
    df = pd.read_excel(excel_path)
    if 'Heads Pruned' in df.columns:
        heads = pd.to_numeric(df['Heads Pruned'], errors='coerce').to_numpy()
        value_df = df.drop(columns=['Heads Pruned'])
    else:
        # 
        heads = np.arange(len(df))
        value_df = df
    box_matrix = []
    for _, row in value_df.iterrows():
        vals = pd.to_numeric(row, errors='coerce')
        vals = vals[vals.notna()].to_numpy(dtype=float)
        box_matrix.append(vals)
    return heads, box_matrix

def load_line_series(csv_path):
    """ Greedy Gnorm 
       - (Pruned_Heads, Accuracy) (x, y)
       -  (None, y)/
    """
    df = pd.read_csv(csv_path)
    df.columns = [str(c).strip() for c in df.columns]
    cols_lower = [c.lower() for c in df.columns]
    if {'pruned_heads','accuracy'}.issubset(set(cols_lower)):
        xcol = df.columns[cols_lower.index('pruned_heads')]
        ycol = df.columns[cols_lower.index('accuracy')]
        x = pd.to_numeric(df[xcol], errors='coerce').to_numpy()
        y = pd.to_numeric(df[ycol], errors='coerce').to_numpy()
        mask = np.isfinite(x) & np.isfinite(y)
        return x[mask], y[mask]
    #  y
    best_col, best_cnt = None, -1
    for c in df.columns:
        ser = pd.to_numeric(df[c], errors='coerce')
        cnt = ser.notna().sum()
        if cnt > best_cnt:
            best_cnt = cnt
            best_col = c
    y = pd.to_numeric(df[best_col], errors='coerce').dropna().to_numpy(dtype=float)
    return None, y

def align_line_every_n(line_x, line_y, step_n, total_heads=None):
    """ step_n 
       -  line_x  x==0..total_heads  step_n 
       -  line_x  y  step_n 
    """
    if line_x is None:
        return line_y[::step_n]
    # line_x  step_n 
    if total_heads is None:
        total_heads = int(np.nanmax(line_x))
    grid = np.arange(0, total_heads + 1, step_n, dtype=int)
    order = np.argsort(line_x)
    x_sorted = np.asarray(line_x)[order]
    y_sorted = np.asarray(line_y)[order]
    # 
    x_unique, idx = np.unique(x_sorted, return_index=True)
    y_unique = y_sorted[idx]
    # 
    y_on_grid = np.interp(grid, x_unique, y_unique, left=y_unique[0], right=y_unique[-1])
    return grid, y_on_grid

def percent_ticks_from_heads(heads, total_heads):
    """ heads """
    perc = np.round((np.asarray(heads)/total_heads)*100).astype(int)
    return perc


# ========== 22  ==========
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("BERT", "ALBERT", "RoBERTa", "XLM-RoBERTa"),
    horizontal_spacing=0.08,
    vertical_spacing=0.12,

)
for ann in fig['layout']['annotations']:
    ann['font'] = dict(size=23, family="Arial")  # size1820


# 
fig.update_layout(
    template="plotly_white",
    width=1200, height=950,
    legend=dict(orientation='h', yanchor='bottom', y=-0.08, xanchor='center', x=0.5),
    margin=dict(l=40, r=20, t=70, b=90),
    # 
    # paper_bgcolor="rgba(0,0,0,0)",
    # plot_bgcolor="rgba(0,0,0,0)",
)

Y_RANGE = [0.5, 1.0]
STEP = 6   #  6 
TOTAL_HEADS_DEFAULT = 144  #  144

def add_panel(fig, row, col, box_path, line_path, model_name, show_legend=False):
    # 1) 
    heads, box_matrix = load_box_excel(box_path)

    # 2) 
    line_x, line_y = load_line_series(line_path)
    total_heads = int(np.nanmax(heads)) if len(heads) else TOTAL_HEADS_DEFAULT

    if line_x is None:
        y_simplified = line_y[::STEP]
        x_simplified = np.arange(0, STEP*len(y_simplified), STEP)
        mask = x_simplified <= total_heads
        x_simplified, y_simplified = x_simplified[mask], y_simplified[mask]
    else:
        x_simplified, y_simplified = align_line_every_n(line_x, line_y, STEP, total_heads)

    if line_x is not None and (line_x == 0).any():
        no_pruning = float(line_y[np.where(line_x == 0)[0][0]])
    else:
        no_pruning = float(line_y[0])

    # 3) 
    fig.add_trace(go.Scatter(x=x_simplified, y=y_simplified,
                             mode='lines', name='Greedy-Gnorm',
                             line=dict(color='green', width=2),
                             legendgroup='grp', showlegend=show_legend),
                  row=row, col=col)

    fig.add_trace(go.Scatter(x=[0, total_heads], y=[no_pruning, no_pruning],
                             mode='lines', name='No Pruning',
                             line=dict(color='orange', width=2, dash='dash'),
                             legendgroup='grp', showlegend=show_legend),
                  row=row, col=col)

    # Random Pruning  + skyblue  + 
    for i, h in enumerate(heads):
        if i >= len(box_matrix):
            break
        fig.add_trace(
            go.Box(
                y=box_matrix[i], x=[h] * len(box_matrix[i]),
                name='Random Pruning',
                boxpoints='outliers',                      # 
                marker=dict(color='red', size=4, opacity=0.9),  # 
                line=dict(color='skyblue', width=1),       # skyblue
                fillcolor='rgba(135,206,235,0.35)',        # skyblue 
                whiskerwidth=0.6,
                legendgroup='grp',
                showlegend=(show_legend and i == 0)        #  legend
            ),
            row=row, col=col
        )

    # 4)  y 
    #  y 
    all_y = []
    all_y.extend(list(y_simplified))
    all_y.append(no_pruning)
    for arr in box_matrix:
        all_y.extend(arr.tolist())
    all_y = [v for v in all_y if np.isfinite(v)]

    min_y, max_y = min(all_y), max(all_y)
    padding = (max_y - min_y) * 0.05  #  5%
    ymin, ymax = min_y - padding, max_y + padding

    # 
    ticks_x = np.arange(0, total_heads+1, STEP*3)
    ticktext = [f"{int(round(x/total_heads*100))}%" for x in ticks_x]

    fig.update_xaxes(title_text="Heads Pruned",
                     tickmode='array', tickvals=ticks_x, ticktext=ticktext,
                     showgrid=True, row=row, col=col)
    fig.update_yaxes(title_text="Accuracy",
                     showgrid=True, range=[ymin, ymax],
                     row=row, col=col)




# ===== =====
add_panel(fig, 1, 1, box_data_path_BERT,   line_data_path_BERT,   "BERT",        show_legend=True)
add_panel(fig, 1, 2, box_data_path_ALBERT, line_data_path_ALBERT, "ALBERT",      show_legend=False)
add_panel(fig, 2, 1, box_data_path_ROBERTA, line_data_path_ROBERTA, "RoBERTa",   show_legend=False)
add_panel(fig, 2, 2, box_data_path_XLM,   line_data_path_XLM,     "XLM-RoBERTa", show_legend=False)

#  traceOutliers Pruning
fig.add_trace(
    go.Scatter(
        x=[None], y=[None],
        mode='markers',
        marker=dict(color='red', size=6),
        name='Outliers Pruning',
        legendgroup='grp',
        showlegend=True
    ),
    row=2, col=2  # 
)

# 
fig.update_layout(title="Accuracy During Heads Pruning")
fig.update_layout(
    font=dict(size=14, family="Arial"),
    title=dict(
        text="Accuracy under Random Head Pruning",  # 
        font=dict(size=27, family="Arial Bold"),   # 
        x=0.5,                                     # 
        xanchor='center'
    )
)

fig.update_layout(
    template="plotly_white",
    width=1600, height=950,
    paper_bgcolor="rgba(0,0,0,0)",
    plot_bgcolor="rgba(0,0,0,0)",
    legend=dict(
        orientation='h',
        yanchor='bottom',
        y=-0.15,        # 
        xanchor='center',
        x=0.5,
        traceorder='normal',
        font=dict(size=22)
    ),
    margin=dict(l=40, r=20, t=70, b=70)  # 
)

fig.show()

#  PNG/PDF kaleidopip install -U kaleido
# fig.write_image("D:/research/research6/figures/random_pruning_grid.png", scale=3)
# fig.write_image("D:/research/research6/figures/random_pruning_grid.pdf")
```
