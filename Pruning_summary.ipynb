```python
# Load model directly
bert_tokenizer = AutoTokenizer.from_pretrained("hw2942/bert-base-chinese-finetuning-financial-news-sentiment-v2")
bert = AutoModelForSequenceClassification.from_pretrained("hw2942/bert-base-chinese-finetuning-financial-news-sentiment-v2",output_attentions=True)

# load the dataset 
bert_ds = load_dataset("hw2942/financial-news-sentiment")
```

```python
def get_acc(ds,model,tokenizer,device='cuda'):
    total = 0
    right = 0
    model.to(device)
    for data in ds['train']:
        inputs = tokenizer(data['Title'],return_tensors='pt').to(device)
        outputs = model(**inputs)
        total  += 1
        if torch.max(outputs[0][0].softmax(0),dim=0).indices==data['labels']:
            right += 1
    return right/total
print('Before pruning:')
print(get_acc(ds=bert_ds, model=bert,tokenizer=bert_tokenizer))
print(summary(bert))
# head_mask,AE_matrix * head_mask,(AE_matrix * head_mask).sum(dim=1)
pruner = TransformerPruner(bert)
pruner.prune(head_mask=head_mask_BERT,save_model=False)
print(head_mask_BERT)
print('After pruning:')
print(get_acc(ds=bert_ds, model=bert,tokenizer=bert_tokenizer))
print(summary(bert))
```

```python
albert_tokenizer = AutoTokenizer.from_pretrained("Alireza1044/albert-base-v2-mnli")
albert = AutoModelForSequenceClassification.from_pretrained("Alireza1044/albert-base-v2-mnli")
albert_ds = load_dataset("nyu-mll/glue", "mnli")
```

```python
def get_acc(ds,model,tokenizer,size=len(albert_ds['validation_matched']),device='cuda'):
    count = 0
    cursor = 0
    model.to(device)
    for data in tqdm(ds['validation_matched']):
        cursor += 1
        inputs = tokenizer(data['premise'],data['hypothesis'],return_tensors='pt').to(device) # 
        with torch.no_grad():
            logits = model(**inputs).logits
        predicted_class_id = logits.argmax().item()
        if predicted_class_id == data['label']:
            count += 1
        if cursor > size:
            break
    return count / size
print('Before pruning:')
print(get_acc(ds=albert_ds, model=albert,tokenizer=albert_tokenizer))
print(summary(albert))
# head_mask,AE_matrix * head_mask,(AE_matrix * head_mask).sum(dim=1)
pruner = TransformerPruner(albert)
pruner.prune(head_mask=head_mask_ALBERT,save_model=False)
print(head_mask_ALBERT)
print('After pruning:')
print(get_acc(ds=albert_ds, model=albert,tokenizer=albert_tokenizer))
print(summary(albert))
```

```python
def get_acc(ds,model,tokenizer,size=len(roberta_ds['validation']),device='cuda'):
    count = 0
    cursor = 0
    model.to(device)
    for data in tqdm(ds['validation']):
        if data['label']==1:
            continue
        cursor += 1
        inputs = tokenizer(data['text'],return_tensors='pt').to(device) # 
        with torch.no_grad():
            predicted_class_id = model(**inputs)
        if torch.max(predicted_class_id[0],dim=1).indices[0] == (1 if data['label']==2 else 0):
            count += 1
        if cursor > size:
            break
    return count / size
print('Before pruning:')
print(get_acc(ds=roberta_ds, model=roberta,tokenizer=roberta_tokenizer,size=1000))
print(summary(roberta))
# head_mask,AE_matrix * head_mask,(AE_matrix * head_mask).sum(dim=1)
pruner = TransformerPruner(roberta)
pruner.prune(head_mask=head_mask_ROBERTA,save_model=False)
print(head_mask_ROBERTA)
print('After pruning:')
print(get_acc(ds=roberta_ds, model=roberta,tokenizer=roberta_tokenizer,size=1000))
print(summary(roberta))
```

```python
def get_acc(ds,model,tokenizer,size=len(xlm_roberta_ds['validation']),device='cuda'):
    count = 0
    id2lang = model.config.id2label
    cursor = 0
    model.to(device)
    for data in ds['validation']:
        cursor += 1
        inputs = tokenizer(data['text'], padding=True, truncation=True, return_tensors="pt").to(device)
        with torch.no_grad():
            logits = model(**inputs).logits
        preds = torch.softmax(logits, dim=-1)
        vals, idx = torch.max(preds, dim=1)
        if id2lang[idx.item()]==data['labels']:
            count += 1
        if cursor > size:
            break
    return count / size
print('Before pruning:')
print(get_acc(ds=xlm_roberta_ds, model=xlm_roberta,tokenizer=xlm_roberta_tokenizer))
print(summary(xlm_roberta))
# head_mask,AE_matrix * head_mask,(AE_matrix * head_mask).sum(dim=1)
pruner = TransformerPruner(xlm_roberta)
pruner.prune(head_mask=head_mask_XLM_ROBERTA,save_model=False)
print(head_mask_XLM_ROBERTA)
print('After pruning:')
print(get_acc(ds=xlm_roberta_ds, model=xlm_roberta,tokenizer=xlm_roberta_tokenizer,size=3000))
print(summary(xlm_roberta))
```
